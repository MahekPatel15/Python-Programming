### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year=int(input("Enter year: "))
if ((year%4==0 and year%100!=0) or (year%400==0)):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
subject1_marks=int(input("Enter marks of subject 1: "))
subject2_marks=int(input("Enter marks of subject 2: "))
subject3_marks=int(input("Enter marks of subject 3: "))
subject4_marks=int(input("Enter marks of subject 4: "))
subject5_marks=int(input("Enter marks of subject 5: "))
percentage=float((subject1_marks+subject2_marks+subject3_marks+subject4_marks+subject5_marks)/5)
print(f"{percentage}")
if percentage < 35:
    grade = 'Fail'
elif percentage >= 35:
    grade = 'Pass class'
elif percentage >= 45:
    grade = 'Second class'
elif percentage >= 60:
    grade = 'First class'
elif percentage >= 70:
    grade = 'Distinction'
    
print(f"Grade obtained: {grade}")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
unit=int(input("Enter units: "))
if unit >=1 and unit<=50:
    price = unit*2.60
elif unit <=100:
    price = (50*2.60) + (unit-50)*3.25
elif unit <=200:
    price = (50*2.60) + (unit-50)*3.25 + (unit-100)*5.26
elif unit > 200:
    price = (50*2.60) + (unit-50)*3.25 +(unit-100)*5.26 + (unit - 200)*8.45
    
print(f"Total price: {price}")
```
