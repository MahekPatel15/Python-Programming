### 03) WAP to accept three integers and display the numbers, their sum, and average.
#### Input: 10 20 30
#### Output:

##### Sum: 60

```python
n1=int(input("Enter n1: "))
n2=int(input("Enter n2: "))
n3=int(input("Enter n3: "))

print(f"Numbers: {n1} {n2} {n3}")
sum = n1+n2+n3
print(f"Sum: {sum}")

average = (n1+n2+n3)/3
print(f"Average: {average}")
```

### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
print("Enter 1 for coverting Degree to Fahrenheit: ")
print("Enter 2 for converting Fahrenheit to Degree: ")
choice = int(input("Enter your choice: "))
match choice:
    case 1:
        c = float(input("Enter temperature in Degree Celsius: "))
        f = (c * 9/5) + 32
        print(f"Temperature in Fahrenheit: {f}")
    case 2:
        f = float(input("Enter temperature in Fahrenheit: "))
        c = (f - 32) * 5/9
        print(f"Temperature in Degree Celsius: {c}")
    case _:
        print("Invalid choice")
```

### 15) WAP to get the distance from user into kilometer, and convert it into meter, feet, inches and centimeter.
